```python
# 台股列表
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from fake_useragent import UserAgent
from datetime import datetime

# 抓取股票資訊
url = "https://histock.tw/stock/rank.aspx?p=all"
user_agent = UserAgent()
headers = {'user-agent': user_agent.random}

res = requests.get(url, headers = headers)
tmp = BeautifulSoup(res.text, 'lxml').select_one('#CPHB1_gv')
df = pd.read_html(tmp.prettify())[0]

# 優化欄位名稱
df.columns = ['stock_no', 'stock_name', 'price', 'ud', 'udp', 'ud_w', 'amp','open', 'high', 'low', 'price_y', 'vol', 'vol_p']
df.head()
```

```python
# 把股票名稱對應股票代號存到json
import json
df_list = df[["stock_no", "stock_name"]]
tw_stock_symbol_code = {}

for i in range(len(df_list)):
    tw_stock_symbol_code[df_list.iloc[i, 1]] = df_list.iloc[i, 0]
tw_stock_symbol_code

with open("data/tw_stock_symbol_code.json", "w", encoding="UTF-8") as f:
    json.dump(tw_stock_symbol_code, f, indent=4)
```

```python
# 檢查 database json file 是否正常
with open('data/tw_stock_symbol_code.json', "r", encoding="UTF-8") as f:
    data = json.load(f)
```
